```python
import os
import pandas as pd
```

```python
metadata_file = '/data/group_data/rag-robust-eval/data/policy/metadata.json'
root_path = '/data/group_data/rag-robust-eval/data/policy/docs'
link_root = 'https://cpd.hud.gov/cpd-public/api/con-plans/downloads/{id}?fileName={file_name}'
```

```python
metadata_list = pd.read_json(metadata_file).to_dict(orient='records')
```

```python
for metadata in metadata_list:
    id = metadata['id']
    file_name = metadata['fileName']
    doc_link = link_root.format(id=id, file_name=file_name)
    
    if os.path.exists(os.path.join(root_path, f"{id}.docx")):
        print(f"File {id}.docx already exists. Skipping download.")
        continue
    # Download the doc file from the link
    os.system(f'wget -O {os.path.join(root_path, f"{id}.docx")} {doc_link}')
```
